```python
import pandas as pd
co2Link='https://github.com/SDSRomina/parcial.tarea/raw/main/CO2%20emissions.xls'
accesToElectricityLink = 'https://github.com/SDSRomina/parcial.tarea/raw/main/Access%20to%20electricity%20.xls'
forestAreaLink = 'https://github.com/SDSRomina/parcial.tarea/raw/main/Forest%20area.xls'

#Aqui tenemos los tres datasframes
co2=pd.read_excel(co2Link)
accesToElectricity = pd.read_excel(accesToElectricityLink)
forestArea = pd.read_excel(forestAreaLink)
```

```python
tresDFS = [co2, accesToElectricity, forestArea]
```

```python
allColumnNames=[]
for df in tresDFS:
    allColumnNames.append(set(df.columns))
allColumnNames
```

```python
commonColumns=set.intersection(*allColumnNames)
nocommonColumns = set.union(*allColumnNames)-commonColumns
allColumns =  commonColumns.union(nocommonColumns)
allColumns     
#columnas_nuevas = set(df.columns)
#columnas_existentes = set().union(*[set(df.columns) for df in lista_de_dataframes])
#columnas = columnas_nuevas && columnas_existentes
```

```python
allDFs_sameNames=[df.loc[:,list(commonColumns)] for df in tresDFS]
allDFs_sameNames
```

```python

co2 = pd.concat([list(commonColumns), list(nocommonColumns)], axis=1)
```

```python
co2.merge(forestArea,how='inner',on = 'Country Code').merge(accesToElectricity, how='inner',on = 'Country Code')
#asignaremos todos esos merge a uno solo
ThreeDfs = co2.merge(forestArea,how='inner',on = 'Country Code').merge(accesToElectricity, how='inner',on = 'Country Code')
```
